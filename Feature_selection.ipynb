```python
import pandas as pd
```

```python
data = pd.read_excel("Data_Video_Volunteers.xlsx")
data.head()
```

```python
cols = list(data.columns)
useful_cols = []
```

```python
sum(data["no_of_individuals_affected"].isna())/len(data)
```

```python
def null_percentages(df):
    columns = list(df.columns)
    null_percents = []
    for column in columns:
       null_percents.append(round((sum(df[column].isna())/len(df))*100, 3))
    return pd.DataFrame(data = zip(columns,null_percents), columns = ['column','null_%'])

null_analysis = null_percentages(data)        
```

```python
# Grouping by null_% and creating categories

null_analysis['range'] = pd.cut(
    null_analysis['null_%'],
    bins=[0, 20, 40, 60, 80, 100],
    labels = [ '<20', '(20,40]', '(40,60]','(60,80]','>80'],
    include_lowest=True)
```

```python
null_analysis = null_analysis.groupby('range')
```

```python
filled = null_analysis.get_group('<20')
filled.sort_values('null_%')
```
